```python
import pandas as pd
df_sales = pd.read_excel("linechart.xlsx")

df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'] , df_sales['Fridge'] , color='red' , label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="orange", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("product sales")
plt.xlabel("Quarter")
plt.ylabel("Million $")

plt.legend()
plt.show()
```

```python
total_sale = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sale
```

```python
total_sale.index
```

```python
plt.pie(total_sale , labels= total_sale.index , autopct="%1.1f%%" , startangle=140, explode=(0.1,0,0) , shadow=True) 
plt.show()
```
