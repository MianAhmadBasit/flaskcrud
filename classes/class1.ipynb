```python
import datetime
class Player:
    def __init__(self,fname,lname , team ,birth):
        self.first_name = fname
        self.last_name = lname
        self.birth_year = birth

    def get_age(self):
        now = datetime.datetime.now()
        return now.year - self.birth_year 

class CricketPlayer(Player):
    def __init__(self,fname , lname ,team,birth ):
        Player.__init__(self,fname , lname ,team,birth)
        self.team = team
        self.scores = []

    def add_scores(self,score):
        self.scores.append(score)

    def get_average_score(self):
        return sum(self.scores)/len(self.scores)
    


class TennisPlayer(Player):
    def __init__(self, fname, lname, team, birth , gwinner):
        Player.__init__(self , fname, team ,lname,  birth)
        self.grand_slam_winner = gwinner
        self.aces=[]

    def get_average_aces_per_match(self):
        return sum(self.aces)/len(self.aces)

    
virat = CricketPlayer('virat' , 'kholi' , 'india' , 1977)
virat.add_scores(55)
virat.add_scores(255)
virat.add_scores(595)

print("The age of virat is " , virat.get_age())
print("The Average score of virat is" , virat.get_average_score())


roger = TennisPlayer("roger" , "ferder" , "usa" ,  1981 , 20)
print("Age of roger feder is" , roger.get_age())




        
```
